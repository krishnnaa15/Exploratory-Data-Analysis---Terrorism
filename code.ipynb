### Cleaning the data

```python
# select specific columns from the DataFrame df and create a new DataFrame with only those columns 

df=df[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
df.head()
```

```python
# Rename the columns in the DataFrame

df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
# Display the first few rows of your DataFrame with the new column names

df.head()
```

```python
# provides a concise summary of the DataFrame, including information about the data types of each column, the number of non-null values, and memory usage

df.info()
```

```python
# Representing the dimensions of the DataFrame [ie.. (row,column)]

df.shape
```

```python
# Count the number of missing values in each column of the DataFrame

df.isnull().sum()
```

```python
df["Killed"]=df["Killed"].fillna(0)
df["Wounded"]=df["Wounded"].fillna(0)
df["Casualty"]=df["Killed"]+df["Wounded"]
```

This above code fills any missing values in the **Killed** and **Wounded** columns with zeros using the **fillna(0)** method and then creates a new column "**Casualty** by adding the values of **Killed** and **Wounded** together. The resulting DataFrame will have the new **Casualty** column with the total casualties.

```python
"""
Used to generate descriptive statistics of the numerical columns in a DataFrame.It provides 
information such as the count, mean, standard deviation, minimum, 25th percentile (Q1), median
(50th percentile or Q2), 75th percentile (Q3), and maximum.
"""

df.describe()
```

***2. Region wise attacks***

+ Distribution of Terrorist Attacks over Regions from 1970-2017

```python
reg=pd.crosstab(df.Year,df.Region)
reg.head()
```

***We are using the crosstab function from the pandas library to create a contingency table (cross-tabulation) for the "Year" and "Region" columns in DataFrame df. This table shows the frequency of occurrences for each combination of year and region.***

In this code:

+ **df.Year** - Specifies the values to be used as rows in the contingency table (in this case, the "Year" column).
+ **df.Region** - Specifies the values to be used as columns in the contingency table (in this case, the "Region" column).
+ **reg.head()** - Displays the first few rows of the resulting contingency table.

```python
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

***We are using Matplotlib to create an area plot based on the contingency table reg that you created using the crosstab function.***

i.Plotting:

reg.plot(kind="area", stacked=False, alpha=0.5, figsize=(20, 10))

+ **kind="area"** - Specifies that an area plot should be created.
+ **stacked=False** - Specifies that the areas should not be stacked on top of each other.
+ **alpha=0.5** - Sets the transparency level of the areas to 0.5 (semi-transparent).
+ **figsize=(20, 10)** - Specifies the size of the figure (width, height) in inches.

ii.Title and Labels:

+ **plt.title("Region-wise Attacks", fontsize=20)** - Sets the title of the plot.

iii.Displaying the Plot:

+ **plt.show()** - Displays the plot.

+ Total Terrorist Attacks in each Region from 1970-2017

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

***We are transposing the contingency table reg using the transpose() function and then calculating the total number of attacks for each region across all years.***

In this code:

+ **regt = reg.transpose()** - Transposes the rows and columns of the contingency table, swapping the "Year" and "Region" roles.
+ **regt["Total"] = regt.sum(axis=1)** - Adds a new column "Total" to regt representing the total number of attacks for each region across all years. The sum(axis=1) calculates the sum across each row.
+ **ra = regt["Total"].sort_values(ascending=False)** - Creates a Series (ra) by sorting the regions based on the total number of attacks in descending order (ascending=False).
+ **ra** - Displays the result, which is a Series showing the regions sorted by the total number of attacks in descending order.

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```
